```python
import pandas as pd
df = pd.read_csv('01_reindeerfood.csv', skip_blank_lines=False, header=None)
```

```python
crew = {}
i = 0
sum = 0
max_food = 0
max_deer = 0
# loop through all rows in the dataframe
for row in df.iterrows():
    # if the row is not empty, add the values up to the sum
    if not pd.isnull(row[1][0]):
        sum += (row[1][0])

    else:
        # if the row is empty, add the sum to the dictionary
        crew[i] = sum
        # if the sum is greater than the max_food, update the max_food and max_deer
        if sum > max_food:
            max_food = sum
            max_deer = i
        # increment the i counter, create new key in dictionary, and reset sum  
        i += 1
        crew[i] = 0
        sum = 0

    
```

```python
print(max_deer, ': ', max_food)
```

```python
crew_sorted = sorted(crew.items(), key=lambda x:x[1], reverse=True)
```
